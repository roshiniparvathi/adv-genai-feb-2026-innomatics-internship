```python
# -----------------------------------
# 1. User Login Check
# -----------------------------------

## Stored credentials
correct_username = "admin"
correct_password = "1234"

# Take input from user
username = input("Enter username: ")
password = input("Enter password: ")

# Check login
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# -----------------------------------
# 2. Pass / Fail Analyzer
# -----------------------------------

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Check each student's marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print result
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)


print("\n-----------------------------\n")
```

```python
# -----------------------------------
# 3. Simple Data Cleaner
# -----------------------------------

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Clean each name
for name in names:
    clean_name = name.strip().lower()   # remove spaces + convert to lowercase
    cleaned_names.append(clean_name)

# Print cleaned list
print("Cleaned Names:", cleaned_names)


print("\n-----------------------------\n")
```

```python
# -----------------------------------
# 4. Message Length Analyzer
# -----------------------------------

messages = ["Hi", "Welcome to the platform", "OK"]

# Check each message
for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)

    # Flag long messages
    if length > 10:
        print("Status: Long Message ⚠️")
    else:
        print("Status: Normal")

    print()


print("-----------------------------\n")
```

```python
# -----------------------------------
# 5. Error Message Detector
# -----------------------------------

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count ERROR entries
for log in logs:
    if log == "ERROR":
        error_count += 1

# Print result
print("Total ERROR Messages:", error_count)
```
